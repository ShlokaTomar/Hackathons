```python
le = LabelEncoder()
y_encoded = le.fit_transform(y)

print(np.unique(y_encoded))  # should be 0 to 11
```

```python
class_counts = np.bincount(y_encoded)
class_weights = 1.0 / class_counts
class_weights = class_weights / class_weights.sum()  # normalize (optional)

class_weights = torch.tensor(class_weights, dtype=torch.float32)
```

```python
X_tensor = torch.tensor(X, dtype=torch.float32)
y_tensor = torch.tensor(y_encoded, dtype=torch.long)
```

```python
class ASLDataset(Dataset):
    def __init__(self, X, y=None):
        self.X = X
        self.y = y

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        if self.y is not None:
            return self.X[idx], self.y[idx]
        return self.X[idx]
```

```python
dataset = ASLDataset(X_tensor, y_tensor)

#A Dataset answers two questions:

#How many samples?

#What is sample i?
```

```python
loader = DataLoader(
    dataset,
    batch_size=32,
    shuffle=True
)
```

```python
class ASLNet(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(63, 64),
           # nn.BatchNorm1d(64),
            nn.ReLU(),
            nn.Dropout(0.15),

            nn.Linear(64, 32),
          #  nn.BatchNorm1d(32),
            nn.ReLU(),
            nn.Dropout(0.15),

            nn.Linear(32, 12)
        )

    def forward(self, x):
        return self.net(x)

```

```python
model = ASLNet()
```

```python
criterion = nn.CrossEntropyLoss()

#crossentropyloss applies softmax internally
```

```python
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
```

```python
epochs = 500

for epoch in range(epochs):
    model.train()
    total_loss = 0

    for X_batch, y_batch in loader:
        optimizer.zero_grad()          # clear old gradients

        logits = model(X_batch)        # forward pass
        loss = criterion(logits, y_batch)

        loss.backward()                # backprop
        optimizer.step()               # update weights

        total_loss += loss.item()

    print(f"Epoch {epoch+1}, Loss: {total_loss/len(loader):.4f}")
```

```python
print("ENCODER CLASSES:", le.classes_)
```

```python
X_test = df_test.drop(columns=["row_ID"], errors="ignore").values
X_test_tensor = torch.tensor(X_test, dtype=torch.float32)

test_dataset = ASLDataset(X_test_tensor)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)
```

```python
model.eval()
predictions = []

with torch.no_grad():
    for X_batch in test_loader:
        logits = model(X_batch)
        preds = torch.argmax(logits, dim=1)
        predictions.extend(preds.cpu().numpy())
```

```python
print(preds[:20])
print(le.inverse_transform(preds[:20]))
```

```python
pred_labels = le.inverse_transform(predictions)
```
